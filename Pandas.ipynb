```python
import pandas as pd
c = pd.DataFrame({'no.':a})
c
```

```python
store = pd.HDFStore('my_data.h5')
```

```python
store['obj1'] = c
store['obj1_col'] = c['no.']
store
```

```python
store['obj1']
```

```python
store['obj1_col']
```

```python
store.put('obj2', c, format='table')
store.select('obj2', where=['index >= 10 and index <= 15'])
store.close()
```
